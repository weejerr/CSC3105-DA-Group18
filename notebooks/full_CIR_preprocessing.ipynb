```python
# load and combine the 7 dataset parts into one DataFrame
import glob
import pandas as pd

files = glob.glob("../data/raw/uwb_dataset_part*.csv")
print("Files found:", len(files), files)

df = pd.concat([pd.read_csv(f) for f in files], ignore_index=True)
print("Dataset shape:", df.shape)

df.head()
```

```python
# Data integrity checks
print("Missing values:", df.isnull().sum().sum())
print("Duplicate rows:", df.duplicated().sum())
print("Class balance:\n", df["NLOS"].value_counts())
```

```python
# use all columns except target label
X = df.drop(columns=["NLOS"])
y = df["NLOS"]

print("Full feature shape:", X.shape)
```

```python
print(X.describe())
```

```python
# scaling 
# standardScalar performs z-score standardization (x - mean) / std
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("Scaled feature shape:", X_scaled.shape)
```

```python
# Train/test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
```
