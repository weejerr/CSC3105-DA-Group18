```python
import glob
import pandas as pd
from pathlib import Path

files = sorted(glob.glob("../data/raw/uwb_dataset_part*.csv"))
df = pd.concat([pd.read_csv(f) for f in files], ignore_index=True)

print("Dataset shape:", df.shape)
df.head()
```

```python
TARGETS = ["NLOS", "RANGE"]

SCALAR_FEATURES = [
    col for col in df.columns
    if (not col.startswith("CIR")) and (col not in TARGETS)
]

print("Number of scalar features:", len(SCALAR_FEATURES))
print("Scalar feature names:", SCALAR_FEATURES)

X = df[SCALAR_FEATURES].copy()
```

```python
idx_path = Path("../data/processed")

train_idx = pd.read_csv(idx_path / "train_idx.csv")["idx"].to_numpy()
test_idx  = pd.read_csv(idx_path / "test_idx.csv")["idx"].to_numpy()

X_train = X.loc[train_idx].reset_index(drop=True)
X_test  = X.loc[test_idx].reset_index(drop=True)

print("X_train shape:", X_train.shape)
print("X_test shape :", X_test.shape)
```

```python
out = Path("../data/processed/scalar")
out.mkdir(parents=True, exist_ok=True)

X_train.to_csv(out / "X_train_scalar.csv", index=False)
X_test.to_csv(out / "X_test_scalar.csv", index=False)

print("Saved:")
print(out / "X_train_scalar.csv")
print(out / "X_test_scalar.csv")
```
