```python
# load and combine the 7 dataset parts into one DataFrame
import glob
import pandas as pd

files = glob.glob("../data/raw/uwb_dataset_part*.csv")
print("Files found:", len(files), files)

df = pd.concat([pd.read_csv(f) for f in files], ignore_index=True)
print("Dataset shape:", df.shape)

df.head()
```

```python
# Data integrity checks
print("Missing values:", df.isnull().sum().sum())
print("Duplicate rows:", df.duplicated().sum())
```

```python
# Separate summary features and CIR waveform features
summary_cols = [
    'RANGE', 'FP_IDX', 'FP_AMP1', 'FP_AMP2', 'FP_AMP3',
    'STDEV_NOISE', 'CIR_PWR', 'MAX_NOISE', 'RXPACC',
    'CH', 'FRAME_LEN', 'PREAM_LEN', 'BITRATE', 'PRFR'
]

X_summary = df[summary_cols]
y = df["NLOS"]
```

```python
cir_cols = [col for col in df.columns if col.startswith("CIR")]
X_cir = df[cir_cols]

print("Summary shape:", X_summary.shape)
print("CIR shape:", X_cir.shape)
```

```python
# Scale CIR before PCA
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_cir_scaled = scaler.fit_transform(X_cir)
```

```python
# Apply PCA to compress CIR features
# keeping 95% variance
from sklearn.decomposition import PCA

pca = PCA(n_components=0.95)
X_cir_pca = pca.fit_transform(X_cir_scaled)

print("Original CIR features:", X_cir.shape[1])
print("Reduced PCA features:", X_cir_pca.shape[1])
```

```python
# Combine summary + PCA into a final feature matrix
import numpy as np

X_final = np.hstack((X_summary.values, X_cir_pca))

print("Final feature shape:", X_final.shape)
```

```python
# Train/test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_final,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
```
